本講座で用いるサンプルデータの読み込みをまとめました。<br>コードを順に実行し結果を確認しましょう。

```python
# Boston house-prices datase for REGRESSION(回帰)
import pandas as pd
from sklearn.datasets import load_boston
from IPython.core.display import display
dataset = load_boston()

# set dataframe
X = pd.DataFrame(dataset.data,
                 columns=dataset.feature_names)
y = pd.DataFrame(dataset.target,
                 columns=['MEDV'])
# check the shape
print('----------------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('y shape: (%i,%i)' %y.shape)
print('-----------------------------------------')
print(y.describe())
display(X.join(y).head(5))
```

```python
print(dataset.DESCR)
```

```python
# big mart sales amount for REGRESSION
# https://datahack.analyticsvidhya.com/contest/practice-problem-big-mart-sales-iii/
import pandas as pd
import numpy as np
df = pd.read_csv('./data/av_big_mart_sales_UWu5bXk.csv',
                 header=0)  # headerなしの場合はNone
X = df.iloc[:,:-1]  # 全行対象,最終カラム以外をXとする
y = df.iloc[:,[-1]] # 全行対象,最終カラムをyとする

# check the shape
print('----------------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('y shape: (%i,%i)' %y.shape)
print('----------------------------------------')
print(y.describe())
# 目的変数に欠損が含まれているとアルゴリズムが学習できないためチェック
print('----------------------------------------')
print('Check null count of target variable: %i' % y.isnull().sum())
print('----------------------------------------')
display(X.join(y).head())
```

```python
# Iris dataset for CLASSIFICATION
# Classic and very easy multi-class classification dataset
import pandas as pd
from sklearn.datasets import load_iris
dataset = load_iris()

# Set dataframe
X = pd.DataFrame(dataset.data,
                 columns=dataset.feature_names)
y = pd.DataFrame(dataset.target,
                 columns=['y'])
# check the shape
print('-----------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('y shape: (%i,%i)' %y.shape)
print('-----------------------------------')
print(y.groupby('y').size())
display(X.join(y).head())
print(dataset.DESCR)
```

```python
# Breast cancer wisconsin dataset for CLASSIFICATION
import pandas as pd
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()

# set dataframe
X = pd.DataFrame(dataset.data,
                 columns=dataset.feature_names)
y = pd.DataFrame(dataset.target,
                 columns=['y'])
# check the shape
print('---------------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('y shape: (%i,%i)' %y.shape)
print('---------------------------------------')
print(y.groupby('y').size())
print('---------------------------------------')
display(X.join(y).head())
# print(dataset.DESCR)
```

```python
# Loan screening data for Classification 
import pandas as pd
df = pd.read_csv('./data/av_loan_u6lujuX_CVtuZ9i.csv',header=0)
X = df.iloc[:, :-1]
y = df.iloc[:, [-1]]

# check the shape
print('-----------------------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('y shape: (%i,%i)' %y.shape)
print('-----------------------------------------------')
print(y.groupby(['Loan_Status']).size())
print('-----------------------------------------------')
print('Check null count of target variable: %i' % y.isnull().sum())
print('-----------------------------------------------')
display(X.join(y).head())

# Converting stirng to number
class_mapping = {'N':1, 'Y':0}
y_new = y.copy()
y_new.loc[:,'Loan_Status'] = y_new['Loan_Status'].map(class_mapping)

# check the result converted
print('-----------------------------------------------')
print('check the result converted')
display(y_new.join(y,rsuffix='_org').head())
print(y_new.groupby(['Loan_Status']).size())
```
