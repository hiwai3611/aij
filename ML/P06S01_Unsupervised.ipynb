# クラスタリング

## KMeansの学習と予測

クラスタリングの学習結果を理解しやすくするため、サンプルデータの生成から始めます。<br>make_blobsは各軸正規分布に従った乱数を生成してくれるデータ生成ライブラリです。

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
%matplotlib inline
import matplotlib.pyplot as plt

X, labels_true = make_blobs(
    n_samples=1000,               # サンプル数
    centers=[[-5,-5],[0,0],[5,5]],# クラスタ重心座標設定(3クラスタを2次元で作成)
    cluster_std=1.0,              # 乱数生成時の標準偏差
    random_state=0)
print('Generated data X:\n', X[:5])
print('Generated data y:\n', labels_true[:5])
```

横軸をX0、縦軸をX1として、生成したサンプルデータを可視化します。<br>クラスタリングによって期待されるのは、データがいずれのクラスタに属するかを正しく予測/表現することです。

```python
df = pd.DataFrame(np.hstack([X,labels_true.reshape(len(labels_true),1)]),
                  columns=["X0","X1","label"])
col = df.label.map({0:'b', 1:'g', 2:'r'})
df.plot(x='X0', y='X1', kind='scatter', c=col, colorbar=False, figsize=(3,3))
```

KMeansの実装です。KMeansではクラスタ数を設定する必要があるため3としました。<br>下記では学習されたクラスタ重心座標を出力しています。<br>先ほど私達が設定した値とほぼ合致することを確認して下さい。

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
print(kmeans.cluster_centers_[0], ':Centroid-0 is a cluster Green')
print(kmeans.cluster_centers_[1], ':Centroid-1 is a cluster Blue')
print(kmeans.cluster_centers_[2], ':Centroid-2 is a cluster Red')
```

学習したモデルを使って未知座標Xのクラスタ番号の予測をしましょう。<br>予測ロジックは単純で予測したいデータと最も近いクラスタ重心座標を持つクラスタ番号を返しているだけです。

```python
kmeans.predict([[-6,-4],[-2,0],[6,8]])
```

### KMeansにおけるクラスタ数の決定

上記では3つの重心座標の周辺にデータを生成したため、私達はデータを3つのクラスタで学習させました。<br>しかし、実際のデータについてはこのような決定はできません。<br>それではクラスタ数をどのように決めたら良いでしょうか。

このような時に役立つのがエルボー法です。クラスタ数を変化させていったとき、クラスタ重心点とクラスタ所属点の間の距離の総和をプロットしたものです。クラスタ数が1であれば近い点もあるでしょうが多くの点は遠方の点になります。またクラスタ数が十分に大きければその総和は0に近づくことが予想できます。<b>エルボー法はこの総和の低下が急速に進む箇所と緩やかに進む箇所、それらを分離する場所でクラスタ数の候補を得ましょうというものです。</b>クラスタ重心点とクラスタ所属点の距離の総和はintertia_で取得できます。

```python
# エルボー図の作成
distortions = []
for i in range(1,11):
    km=KMeans(n_clusters=i, random_state=0).fit(X)
    distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o')
```

上記の通り、クラスタ数が3まではクラスタ数を増やすことで重心点とサンプル間の距離が急速に減少します。しかしクラスタを4以上にしても、もはやその低下幅は限定的です。とすると、このデータについてはクラスタ数は3で十分ではないかと考えられるわけです。

## DBSCANによる学習と予測

最後に密度ベースのクラスタリングであるDBSCANの実装について学びます。<br>ここではDBSCANのご利益を確認すべくサンプルデータを変更します。<br>make_moonsというライブラリで月型のデータサンプルを生成します。

```python
from sklearn.datasets import make_moons
X, labels_true = make_moons(
    n_samples=1000,
    noise=0.05,
    random_state=0)
print('Generated Data:\n', X[:5])
print('\nGenerated Label:\n', labels_true[:5])
```

以下で新しい生成サンプルデータの散布図を描いています。<br>色は本データの理想的なクラスタ分割のされ方です。

```python
df = pd.DataFrame(np.hstack([X, labels_true.reshape(len(labels_true),1)]),
                  columns=["X0","X1","label"])
col = df.label.map({0:'g', 1:'b'})
df.plot(x='X0', y='X1', kind='scatter', c=col, colorbar=False, figsize=(6,4))
```

さて、上記のデータに対してKmeansを学習するとどうなるでしょうか。<br>最適クラスタ数は2とわかっているのでクラスタ数を2として学習させ、クラスタ重心座標を出力します。

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print(kmeans.cluster_centers_[0])
print(kmeans.cluster_centers_[1])
```

クラスタ重心座標の位置が、各月で囲まれた中心領域に存在しています。<br>実際にこの学習データXをどのようにクラスタ化したか表示して確認してみます。

```python
fig1, ax1 = plt.subplots()
for j in range(3):
    ax1.plot(X[kmeans.predict(X) == j, 0],
             X[kmeans.predict(X) == j, 1], 'o',
             label='series ' + str(j))
```

Kmeansの予測は期待通りでしょうか。おそらく違うはずです。距離ベースのアルゴリズムのため、データ空間の右側と左側でクラスタが定義されてしまったようです。もちろん、距離ベースでも密度の連なりをある程度表現できるでしょう。しかし、このように表現しきれないケースも存在するということを把握しましょう。さてこのデータをDBSCANで学習するとどうなるでしょうか。

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
fig2, ax2 = plt.subplots()

# 学習と予測を同時に実施
for j in range(3):
    ax2.plot(X[db.fit_predict(X) == j, 0],
             X[db.fit_predict(X) == j, 1], 'o',
             label='series ' + str(j))
```

上記の通り、このような形状にはDBSCANがよりフィットしていることがわかりました。実際のデータはより高次元でしょうからシンプルにKmeansが良い、DBSCANが良いと事前に決定することは困難です。クラスタ別の特徴や、クラスタ別に実施されたモデリングや施策効果結果等を通し、時に主観的、時に実証的に判断していく必要があることを覚えておきましょう。
